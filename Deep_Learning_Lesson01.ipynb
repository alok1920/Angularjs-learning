# CHUNK 1: Install & Import (Run this first)

```python
import torch
import numpy as np

print("Pytorch version: ", torch.__version__)
print("We are ready to explore tensor!")
```

# CHUNK 2: Create Your First Tensors

```python
# SCALAR (0D TENSOR) - JUST ONE NUMBER ----
scalar = torch.tensor(42)
print("SCALAR:", scalar)
print("Shape:", scalar.shape)
print("Dimensions:", scalar.ndim) # 0
print("─" * 40)

# VECTOR (1D tensor) - a list of numbers ---
vector = torch.tensor([12, 45, 67, 23, 89])
print("VECTOR:", vector)
print("Shape:", vector.shape) # torch.Size([5]) = 5 numbers
print("Dimensions:", vector.ndim) # 1
print("─" * 40)

# MATRIX (2D tensor) -- a table of numbers ---
matrix = torch.tensor([
[12, 45, 67],
[23, 11, 89],
[45, 67, 12]
])
print("MATRIX:\n", matrix)
print("Shape:", matrix.shape)
print("Dimensions:", matrix.ndim) # 2
```

# CHUNK 3: The SHAPE Concept (Most Important!)

```python
# SHAPE tells you the SIZE of each dimension
# Think of it as: shape = (depth, rows, columns)

# Let's create a FAKE COLOR IMAGE (3D tensor)
fake_image = torch.zeros(3,244,244)
# zeros = fill with 0s for now

print("Fake Image Tensor Shape:", fake_image.shape) #shape = 3,244,244
print("What each number means:")
print(f" → {fake_image.shape[0]} color channels (R, G, B)") # 3
print(f" → {fake_image.shape[1]} pixels tall") # 244
print(f" → {fake_image.shape[2]} pixels wide") # 244
print(f" → Total numbers in this image: {fake_image.shape[0] * fake_image.shape[1] * fake_image.shape[2]:,}") # 150,528

print("\n─── Now a BATCH of 32 images ───")
batch_of_images = torch.zeros(32, 3, 224, 224)
print("Batch Shape:", batch_of_images.shape)
print("What each number means:")
print(f" → {batch_of_images.shape[0]} images in this batch")
print(f" → {batch_of_images.shape[1]} color channels each")
print(f" → {batch_of_images.shape[2]} x {batch_of_images.shape[3]} pixels each")
print(f" → Total numbers:  {32 * 3 * 224 * 224:,}")
```

# CHUNK 4: NumPy vs PyTorch (Connecting What You Know!)

```python
# You already know NumPy!
# PyTorch tensors are like NumPy arrays BUT
# they can run on GPU and track gradients (for learning)

# NumPy array you already know
numpy_array = np.array([1.0, 2.0, 3.0, 4.0])
print("NumPy array:", numpy_array)
print("NumPy type:", type(numpy_array))

# Converting to PyTorch tensor (super easy!)
pytorch_tensor = torch.from_numpy(numpy_array)
print("\nPyTorch tensor:", pytorch_tensor)
print("PyTorch type:", type(pytorch_tensor))

# They share the same data!
print("\n─── They behave the same way ───")
print("NumPy mean:", numpy_array.mean())
print("PyTorch mean:", pytorch_tensor.mean())

# The BIG difference - PyTorch can go to GPU!
print("\n─── The superpower PyTorch has ───")
print("Can use GPU:", torch.cuda.is_available())
print("(GPU makes training 10-100x faster!)")
```

# CHUNK 5: Real Example - Text Becomes Tensor

```python
# How does a sentence become numbers?
# This is the foundation of ALL LLMs (ChatGPT etc)

sentence = "I love AI"
words = sentence.split()
print("Original sentence:", sentence)
print("Split into words:", words)

# Step 1: Build a vocabulary (word → number mapping)
vocabulary = {"I": 0, "love": 1, "AI": 2,
              "hate": 3, "coding": 4}

# Step 2: Convert words to numbers
word_numbers = [vocabulary[word] for word in words]
print("\nWords to numbers:", word_numbers)

# Step 3: Make it a tensor!
text_tensor = torch.tensor(word_numbers)
print("Text as tensor:", text_tensor)
print("Shape:", text_tensor.shape)
print(f"\n'{sentence}' is now just numbers: {text_tensor.tolist()}")
print("THIS is how ChatGPT reads your text!")
```
